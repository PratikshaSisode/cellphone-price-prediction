# Split into features and labels

```python
X = train.drop('price_range',axis=1)
y = train['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train1 = sc.transform(X_train)
X_test1 = sc.transform(X_test)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train , Y_train)
```

```python
pred = dtc.predict(X_test)
pred
```

```python
Y_test
```

# Confusion Matrix

```python
from sklearn.metrics import accuracy_score, confusion_matrix
dtc_acc = accuracy_score(Y_test,pred)
print(dtc_acc)
print(confusion_matrix(pred,Y_test))
```

# SVC 

```python
from sklearn.svm import SVC
knn=SVC()
knn.fit(X_train,Y_train)
```

```python
pred1 = knn.predict(X_test)
pred1
```

```python
from sklearn.metrics import accuracy_score
svc_acc = accuracy_score(pred1,Y_test)
print(svc_acc)
print(confusion_matrix(pred1,Y_test))
```

```python
knn.score(X_test,Y_test)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=100, criterion='gini')
classifier.fit(X_train, Y_train)
```

```python
classifier.score(X_test, Y_test)
```

```python
classifier_acc=classifier.score(X_test,Y_test)
```

```python
plt.figure(figsize=(8,6))
model_acc = [dtc_acc,svc_acc,classifier_acc]
model_name = ['DecisionTreeClassifier','SVC','RandomForestClassifier']
sns.barplot(x=model_acc, y=model_name,palette='magma')
plt.xlabel("Accuracy Score")
plt.ylabel("Algorithms")
plt.show()
```
